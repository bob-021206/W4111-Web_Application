# W4111 Task 6 — classicmodels API & layers (`app/main.py`)

This notebook satisfies **Task 6**: it exercises **`classicmodels`** routes (**`/customers`**, **`/orders`**, **`/orderdetails`**) over HTTP with **`requests`** (same layout as `harry_potter_paths.ipynb`). A later section calls **`CustomerResource`**, **`OrderResource`**, and **`OrderDetailsResource`** directly to demonstrate **Tasks 1–4** (`MySQLDataService` + resources). Optional cells at the end mirror **`harry_potter_paths.ipynb`**: mutating examples and **`/harry-potter`** (JSON file).

**Prerequisite:** Load MySQL **`classicmodels`** and create **`.env`** (see **`.env.example`**). Run the API from the repo root, e.g. **`python -m app.main`** or **`uvicorn app.main:app --reload --port 8000`**. Default base URL **`http://127.0.0.1:8000`** — override with **`API_BASE_URL`** if needed.

**Writes:** **`POST`**, **`PUT`**, **`DELETE`** cells change MySQL or JSON data. Run mutating cells cautiously on a shared database.

```python
import os

import requests

BASE_URL = os.environ.get("API_BASE_URL", "http://127.0.0.1:8000").rstrip("/")

try:
    _health = requests.get(f"{BASE_URL}/health", timeout=5)
    _health.raise_for_status()
except requests.RequestException as exc:
    raise RuntimeError(
        f"Cannot reach API at {BASE_URL!r}. Start the server from the repo root, then rerun this cell."
    ) from exc

print("OK:", BASE_URL)
```

## `GET /customers` — list / filter (equality query parameters)

```python
resp = requests.get(f"{BASE_URL}/customers", timeout=30)
assert resp.status_code == 200, resp.text
items = resp.json()["items"]
assert len(items) > 0
SAMPLE_CUSTOMER_NUMBER = items[0]["customerNumber"]
items[:2]
```

## `GET /customers/{customerNumber}` — read one

```python
r = requests.get(f"{BASE_URL}/customers/{SAMPLE_CUSTOMER_NUMBER}", timeout=30)
assert r.status_code == 200
r.json()
```

## `GET /customers/{customerNumber}` — `404` when missing

```python
missing = requests.get(f"{BASE_URL}/customers/99999999", timeout=30)
assert missing.status_code == 404
missing.json()
```

## `GET /orders` and `GET /orders/{orderNumber}`

```python
resp = requests.get(f"{BASE_URL}/orders", timeout=30)
assert resp.status_code == 200
orders = resp.json()["items"]
assert len(orders) > 0
SAMPLE_ORDER_NUMBER = orders[0]["orderNumber"]
SAMPLE_ORDER_NUMBER
```

```python
r = requests.get(f"{BASE_URL}/orders/{SAMPLE_ORDER_NUMBER}", timeout=30)
assert r.status_code == 200
r.json()
```

## `GET /orderdetails` and lines for one order

```python
resp = requests.get(
    f"{BASE_URL}/orderdetails",
    params={"orderNumber": SAMPLE_ORDER_NUMBER},
    timeout=30,
)
assert resp.status_code == 200
lines = resp.json()["items"]
assert len(lines) > 0
SAMPLE_PRODUCT_CODE = lines[0]["productCode"]
lines[0]
```

```python
r = requests.get(
    f"{BASE_URL}/orders/{SAMPLE_ORDER_NUMBER}/orderdetails",
    timeout=30,
)
assert r.status_code == 200
len(r.json()["items"])
```

```python
r = requests.get(
    f"{BASE_URL}/orders/{SAMPLE_ORDER_NUMBER}/orderdetails/{SAMPLE_PRODUCT_CODE}",
    timeout=30,
)
assert r.status_code == 200
r.json()
```

## Tasks 1–4 — direct `Resource` / `MySQLDataService` (same `.env` as the API)

Loads **`.env`** from the project root and imports **`CustomerResource`**, **`OrderResource`**, **`OrderDetailsResource`**. Run Jupyter with the working directory set to the repo root or **`tests/`** (the cell resolves **`app/main.py`** automatically).

```python
import sys
from pathlib import Path

try:
    from dotenv import load_dotenv
except ImportError:
    load_dotenv = None

_cwd = Path.cwd().resolve()
if (_cwd / "app" / "main.py").exists():
    PROJECT_ROOT = _cwd
elif (_cwd.parent / "app" / "main.py").exists():
    PROJECT_ROOT = _cwd.parent
else:
    PROJECT_ROOT = _cwd

sys.path.insert(0, str(PROJECT_ROOT))
if load_dotenv:
    load_dotenv(PROJECT_ROOT / ".env")

from app.resources.CustomerResource import CustomerResource
from app.resources.OrderResource import OrderResource
from app.resources.OrderDetailsResource import OrderDetailsResource

cr, or_, odr = CustomerResource(), OrderResource(), OrderDetailsResource()
cc = cr.get({})
oc = or_.get({})
odc = odr.get({})
print("customers:", len(cc.items), "orders:", len(oc.items), "orderdetails:", len(odc.items))
cc.items[0].model_dump()
```

## Mutating examples (optional)

Creates a throwaway customer, updates it, then deletes it (same pattern as **`harry_potter_paths.ipynb`**). Omit **`salesRepEmployeeNumber`** unless it matches an existing **`employees.employeeNumber`** (foreign key).

```python
payload = {
    "customerName": "Notebook Test Co",
    "contactLastName": "Test",
    "contactFirstName": "Notebook",
    "phone": "555-0100",
    "addressLine1": "1 Notebook Lane",
    "city": "NYC",
    "country": "USA",
}
resp = requests.post(f"{BASE_URL}/customers", json=payload, timeout=30)
assert resp.status_code == 200, resp.text
new_cust_id = int(resp.json())
new_cust_id
```

```python
body = {
    "customerName": "Notebook Test Co Updated",
    "contactLastName": "Test",
    "contactFirstName": "Notebook",
    "phone": "555-0100",
    "addressLine1": "1 Notebook Lane",
    "city": "NYC",
    "country": "USA",
}
r = requests.put(f"{BASE_URL}/customers/{new_cust_id}", json=body, timeout=30)
assert r.status_code == 200 and r.json() == {"updated": 1}
requests.get(f"{BASE_URL}/customers/{new_cust_id}", timeout=30).json()
```

```python
r = requests.delete(f"{BASE_URL}/customers/{new_cust_id}", timeout=30)
assert r.status_code == 200 and r.json()["deleted"] == 1
gone = requests.get(f"{BASE_URL}/customers/{new_cust_id}", timeout=30)
assert gone.status_code == 404
```
